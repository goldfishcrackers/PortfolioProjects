```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#there was an error opening it, adding encoding to make sure the file would open
seoul = pd.read_csv("SeoulBikeData.csv", encoding='unicode_escape')
seoul.head(100)
```

```python
#compare the average # of bikes rented by the time of day
#hours: morning 6 - 12
# afternoon 12 - 18
# evening 18- 24 0 - 6

#lists for designated group
morning_winter = []
afternoon_winter = []
evening_winter = []

morning_fall = []
afternoon_fall = []
evening_fall = []

morning_spring = []
afternoon_spring = []
evening_spring = []

morning_summer = []
afternoon_summer = []
evening_summer = []

#organzie by hour + season
for i in range(len(seoul)):
    x = seoul['Hour'][i]
    y = seoul['Seasons'][i]
    
    #morning
    if (x >= 6 and x <= 12):
        if (y == 'Winter'):
            morning_winter.append(seoul['Rented Bike Count'][i])
        elif (y == 'Summer'):
            morning_summer.append(seoul['Rented Bike Count'][i])
        elif (y == 'Spring'):
            morning_spring.append(seoul['Rented Bike Count'][i])
        elif (y == 'Autumn'):
            morning_fall.append(seoul['Rented Bike Count'][i])
    #start with afternoon
    if x > 12 and x <= 18:
        if (y == 'Winter'):
            afternoon_winter.append(seoul['Rented Bike Count'][i])
        elif (y == 'Summer'):
            afternoon_summer.append(seoul['Rented Bike Count'][i])
        elif (y == 'Spring'):
            afternoon_spring.append(seoul['Rented Bike Count'][i])
        elif (y == 'Autumn'):
            afternoon_fall.append(seoul['Rented Bike Count'][i])
    
    #for the evening
    elif x == 0 and x < 6:
        if (y == 'Winter'):
            evening_winter.append(seoul['Rented Bike Count'][i])
        elif (y == 'Summer'):
            evening_summer.append(seoul['Rented Bike Count'][i])
        elif (y == 'Spring'):
            evening_spring.append(seoul['Rented Bike Count'][i])
        elif (y == 'Autumn'):
            evening_fall.append(seoul['Rented Bike Count'][i])
    elif x > 18 and x <= 24:
        if (y == 'Winter'):
            evening_winter.append(seoul['Rented Bike Count'][i])
        elif (y == 'Summer'):
            evening_summer.append(seoul['Rented Bike Count'][i])
        elif (y == 'Spring'):
            evening_spring.append(seoul['Rented Bike Count'][i])
        elif (y == 'Autumn'):
            evening_fall.append(seoul['Rented Bike Count'][i])

```

```python
#lets average all these values

# averages for morning
avg_morning = []
avg_morning.append(np.average(morning_winter))
avg_morning.append(np.average(morning_fall))
avg_morning.append(np.average(morning_summer))
avg_morning.append(np.average(morning_spring))


#averages for afternoon
avg_aft = []
avg_aft.append(np.average(afternoon_winter))
avg_aft.append(np.average(afternoon_fall))
avg_aft.append(np.average(afternoon_summer))
avg_aft.append(np.average(afternoon_spring))

#averages for evening
avg_eve = []
avg_eve.append(np.average(evening_winter))
avg_eve.append(np.average(evening_fall))
avg_eve.append(np.average(evening_summer))
avg_eve.append(np.average(evening_spring))
```

```python
#lets plot those averages
plt.style.use('bmh')

fig, ax = plt.subplots()

#fun parameters
time_szn = ['Winter', 'Fall', 'Spring', 'Summer']
color = ['#a8e6cf', '#ffd3b6', '#dcedc1', '#ffaaa5']
bar_labels = ['Winter', 'Fall', 'Spring', 'Summer']

#bar colors
ax.bar(time_szn, avg_morning, label = bar_labels, color = color)
ax.set_title('Average amount of bike rides for morning for each season')

```

```python
#make a plot for relationship b/t temp and
# of bikes differentiate between seasons w/ the plot

#arrays for rented bike count vs temp
bikes_winter = np.zeros([2, 2161])
bikes_spring = np.zeros([2, 2207])
bikes_fall = np.zeros([2, 2183])
bikes_summer = np.zeros([2, 2269])

for i in range(len(seoul)):
    x = seoul['Seasons'][i]
    if x == 'Winter':
        bikes_winter[0][i] = seoul['Rented Bike Count'][i]
        bikes_winter[1][i] = seoul['Temperature (°C)'][i]
    elif x == 'Spring':
        bikes_spring[0][i - 2208] = seoul['Rented Bike Count'][i]
        bikes_spring[1][i - 2208] = seoul['Temperature (°C)'][i]
    elif x == 'Autumn':
        bikes_fall[0][i - 6578] = seoul['Rented Bike Count'][i]
        bikes_fall[1][i - 6578] = seoul['Temperature (°C)'][i]
    elif x == 'Summer':
        bikes_summer[0][i - 4370] = seoul['Rented Bike Count'][i]
        bikes_summer[1][i - 4370] = seoul['Temperature (°C)'][i]
```

```python
fig, ax = plt.subplots()

# I added all of them on top of each other to see some differences
ax.scatter(bikes_winter[0],bikes_winter[1], label = 'Winter', color = '#a8e6cf')
ax.scatter(bikes_spring[0],bikes_spring[1], label = 'Spring', color = '#dcedc1')
ax.scatter(bikes_fall[0],bikes_fall[1], label = 'Fall', color = '#ffd3b6')
ax.scatter(bikes_summer[0],bikes_summer[1], label = 'Summer', color = '#ffaaa5')

ax.set_ylabel('Temperature (°C)')
ax.set_xlabel('Rented Bike Count')
ax.legend()
ax.set_title('Rented Bike Count per day vs Temperature - All Seasons')

plt.show()
```
